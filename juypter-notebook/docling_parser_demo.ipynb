## JSON 기반 Segmentation (실전 방법)

Docling의 **label만 믿지 말고** label + reading order + bbox + text rule을 함께 써야 안정적입니다.

- **방식 1. Top-down section segmentation**: `body.children` 순서대로 순회 후 anchor 기반 구간 자르기 (section_header, table, "청구금액", "입금 계좌", page_footer 등).
- **방식 2. Y축 band**: bbox의 top/bottom으로 수평 구간(issuer header / meta / table / summary / footer) 나누기.
- **방식 3. Semantic**: key_value_area 안에서 텍스트 의미로 document_meta, party_meta, shipping_meta, cargo_meta 등 subsegment 분해.

아래 코드는 이 문서(Invoice_25)에 맞춘 **anchor + reading order** 기반 최종 segmentation 예시입니다.

```python
import json


# ----- 1) $ref 해석 헬퍼 (label + text + bbox 활용용) -----
def _ref_to_key(ref):
    """'#/texts/0' -> ('texts', 0), '#/groups/1' -> ('groups', 1)"""
    if not isinstance(ref, dict) or "$ref" not in ref:
        return None
    path = ref["$ref"].strip("#/")
    part, idx = path.rsplit("/", 1)
    return (part, int(idx))


def _get_item(doc_dict, key):
    part, idx = key
    return doc_dict.get(part, [])[idx] if part in doc_dict else None


def get_label_and_text(doc_dict, key):
    item = _get_item(doc_dict, key)
    if item is None:
        return None, None
    label = item.get("label")
    text = (item.get("text") or item.get("orig") or "").strip()
    return label, text


# ----- 2) body.children reading order 확인 (Step 1) -----
body = doc_dict.get("body") or {}
children = body.get("children") or []
print("body.children 순서 (reading order):")
for i, ref in enumerate(children):
    key = _ref_to_key(ref)
    if key:
        label, text = get_label_and_text(doc_dict, key)
        ref_str = f"{key[0]}/{key[1]}"
        print(f"  {i}: {ref_str}  label={label}  text={ (text or '')[:50] }")

# ----- 3) 이 문서(Invoice_25)에 맞는 최종 segmentation 결과 (anchor 기반) -----
# anchor: INVOICE → document_title, table → charge_table, 청구금액 → invoice_total,
#         입금 계좌 번호 → payment_instruction, page_footer → footer
segmentation_result = {
    "segments": [
        {"type": "issuer_header", "items": ["texts/0", "groups/0"]},
        {"type": "document_title", "items": ["texts/3"]},
        {
            "type": "invoice_meta",
            "items": ["groups/1"],
            "subsegments": [
                "document_meta",
                "party_meta",
                "shipping_meta",
                "cargo_meta",
            ],
        },
        {"type": "charge_table", "items": ["tables/0"]},
        {"type": "invoice_total", "items": ["texts/45", "texts/46", "texts/47"]},
        {"type": "contact_info", "items": ["groups/2"]},
        {"type": "payment_instruction", "items": ["texts/51", "groups/3"]},
        {"type": "footer", "items": ["texts/56"]},
    ]
}

print("\n최종 segmentation (anchor + reading order):")
print(json.dumps(segmentation_result, ensure_ascii=False, indent=2))
display(JSON(segmentation_result))
```

```python
def resolve_ref(doc: dict, ref: str) -> dict:
    """
    지원:
    - '#/texts/0'
    - 'texts/0'
    """
    if ref.startswith("#/"):
        ref = ref[2:]
    elif ref.startswith("/"):
        ref = ref[1:]

    parts = ref.split("/")
    cur = doc

    for p in parts:
        if p.isdigit():
            cur = cur[int(p)]
        else:
            cur = cur[p]

    return cur


def get_ref_type(ref: str) -> str:
    ref = ref.lstrip("#/")

    if ref.startswith("texts/"):
        return "text"
    if ref.startswith("groups/"):
        return "group"
    if ref.startswith("tables/"):
        return "table"
    return "unknown"


def print_group_contents(doc: dict, group_obj: dict) -> None:
    print(f"GROUP label={group_obj.get('label')} self_ref={group_obj.get('self_ref')}")
    for child in group_obj.get("children", []):
        ref = child["$ref"]
        item = resolve_ref(doc, ref)
        print(f"  - ref={ref}, label={item.get('label')}, text={item.get('text', '')}")


def print_table_contents(table_obj: dict) -> None:
    print(f"TABLE self_ref={table_obj.get('self_ref')}")
    grid = table_obj["data"]["grid"]

    for row_idx, row in enumerate(grid):
        row_values = [cell.get("text", "") for cell in row]
        print(f"row[{row_idx}] = {row_values}")


def print_segment(doc: dict, segment: dict) -> None:
    print("\n" + "=" * 60)
    print(f"SEGMENT TYPE: {segment['type']}")
    print("=" * 60)

    if "subsegments" in segment:
        print(f"SUBSEGMENTS: {segment['subsegments']}")

    for ref in segment["items"]:
        obj = resolve_ref(doc, ref)
        label = obj.get("label")
        ref_type = get_ref_type(ref)

        if ref_type == "text":
            print(f"[TEXT] ref={ref}, label={label}, text={obj.get('text', '')}")

        elif ref_type == "group":
            print(f"[GROUP] ref={ref}, label={label}")
            print_group_contents(doc, obj)

        elif ref_type == "table":
            print(f"[TABLE] ref={ref}, label={label}")
            print_table_contents(obj)

        else:
            print(f"[UNKNOWN] ref={ref}, obj={obj}")


def print_all_segments(doc: dict, segments: list[dict]) -> None:
    for segment in segments:
        print_segment(doc, segment)


# doc = load_doc(doc_dict)
print(segmentation_result)
# print_body_children(doc_dict)
print_all_segments(doc_dict, segmentation_result["segments"])
```

```python
from collections import defaultdict


def normalize_ref(ref: str) -> str:
    return ref[2:] if ref.startswith("#/") else ref


def resolve_ref(doc: dict, ref: str) -> dict:
    ref = normalize_ref(ref)
    cur = doc
    for p in ref.split("/"):
        cur = cur[int(p)] if p.isdigit() else cur[p]
    return cur


def get_bbox(obj: dict):
    prov = obj.get("prov", [])
    if not prov:
        return None
    return prov[0]["bbox"]


def get_text_items_from_group(doc: dict, group_ref: str):
    group = resolve_ref(doc, group_ref)
    items = []
    for child in group["children"]:
        obj = resolve_ref(doc, child["$ref"])
        bbox = get_bbox(obj)
        if bbox:
            items.append({
                "text": obj.get("text", "").strip(),
                "label": obj.get("label"),
                "bbox": bbox
            })
    return items


def cluster_rows(items, y_tol=12):
    # BOTTOMLEFT 기준에서는 top(t) 값이 비슷하면 같은 행으로 봄
    items = sorted(items, key=lambda x: (-x["bbox"]["t"], x["bbox"]["l"]))
    rows = []

    for item in items:
        placed = False
        y = item["bbox"]["t"]
        for row in rows:
            if abs(row["y"] - y) <= y_tol:
                row["items"].append(item)
                row["ys"].append(y)
                row["y"] = sum(row["ys"]) / len(row["ys"])
                placed = True
                break
        if not placed:
            rows.append({"y": y, "ys": [y], "items": [item]})

    for row in rows:
        row["items"] = sorted(row["items"], key=lambda x: x["bbox"]["l"])
    return rows


def split_columns(rows, page_mid_x=297):
    left_rows, right_rows = [], []

    for row in rows:
        left_items = [it for it in row["items"] if it["bbox"]["l"] < page_mid_x]
        right_items = [it for it in row["items"] if it["bbox"]["l"] >= page_mid_x]

        if left_items:
            left_rows.append(left_items)
        if right_items:
            right_rows.append(right_items)

    return left_rows, right_rows


def row_to_text(row_items):
    return " ".join([it["text"] for it in row_items if it["text"]]).strip()


def parse_kv_rows(row_groups, known_keys):
    results = []
    current = None

    for row_items in row_groups:
        line = row_to_text(row_items)
        if not line:
            continue

        matched_key = None
        for key in known_keys:
            if line.startswith(key):
                matched_key = key
                break

        if matched_key:
            value = line[len(matched_key):].strip()
            value = value.lstrip(":").strip()
            current = {"key": matched_key, "value": value}
            results.append(current)
        else:
            # continuation line
            if current:
                current["value"] = (current["value"] + " " + line).strip()

    return results
```

```python
def resolve_ref(doc: dict, ref: str) -> dict:
    ref = ref.lstrip("#/")
    cur = doc
    for p in ref.split("/"):
        if p.isdigit():
            cur = cur[int(p)]
        else:
            cur = cur[p]
    return cur


def extract_text_items(doc: dict, group_ref: str):
    group = resolve_ref(doc, group_ref)

    items = []

    for child in group["children"]:
        obj = resolve_ref(doc, child["$ref"])

        if obj.get("text"):
            prov = obj.get("prov", [])
            if not prov:
                continue

            bbox = prov[0]["bbox"]

            items.append({"text": obj["text"].strip(), "bbox": bbox})

    return items


def cluster_rows(items, y_tol=12):

    items = sorted(items, key=lambda x: (-x["bbox"]["t"], x["bbox"]["l"]))

    rows = []

    for item in items:

        y = item["bbox"]["t"]
        placed = False

        for row in rows:

            if abs(row["y"] - y) <= y_tol:
                row["items"].append(item)
                row["ys"].append(y)
                row["y"] = sum(row["ys"]) / len(row["ys"])
                placed = True
                break

        if not placed:
            rows.append({"y": y, "ys": [y], "items": [item]})

    for row in rows:
        row["items"] = sorted(row["items"], key=lambda x: x["bbox"]["l"])

    return rows


def split_columns(rows, page_mid_x=300):

    left_rows = []
    right_rows = []

    for row in rows:
        left = []
        right = []

        for item in row["items"]:
            x = item["bbox"]["l"]

            if x < page_mid_x:
                left.append(item)
            else:
                right.append(item)

        if left:
            left_rows.append(left)

        if right:
            right_rows.append(right)

    return left_rows, right_rows


def row_to_text(row):

    parts = []

    for item in row:
        t = item["text"].strip()

        if not t:
            continue

        parts.append(t)

    line = " ".join(parts)

    # ":" 앞뒤 정리
    line = line.replace(" :", ":")
    line = line.replace(": ", ":")

    return line.strip()


def parse_rows(rows, anchors):

    results = {}
    current_key = None

    for row in rows:

        line = row_to_text(row)

        if not line:
            continue

        found_key = None

        for key in anchors:

            if line.startswith(key):

                found_key = key
                break

        if found_key:

            parts = line.split(":", 1)

            if len(parts) == 2:
                value = parts[1].strip()
            else:
                value = ""

            current_key = found_key

            results[current_key] = value

        else:

            # continuation line
            if current_key:

                results[current_key] += " " + line

    return results


def parse_key_values(rows, anchors):

    results = []
    current = None

    for row in rows:

        line = row_to_text(row)

        if not line:
            continue

        matched_key = None

        for key in anchors:
            if line.startswith(key):
                matched_key = key
                break

        if matched_key:

            value = line[len(matched_key) :].strip()
            value = value.lstrip(":").strip()

            current = {"key": matched_key, "value": value}

            results.append(current)

        else:
            if current:
                current["value"] += " " + line

    return results


FIELD_MAP = {
    "INVOICE NO": "invoice_no",
    "B/L NO": "bl_no",
    "SHIPPER": "shipper",
    "VESSEL/VOY": "vessel_voy",
    "P.O.L.": "pol",
    "P.O.D.": "pod",
    "WEIGHT": "weight",
    "MEASUREMENT": "measurement",
    "MRN": "mrn",
    "MSN/HSN": "msn_hsn",
    "출력일자": "print_date",
    "PKG'S": "packages",
    "TERM": "term",
}


def normalize_fields(kv_list):

    result = {}

    for item in kv_list:

        key = item["key"]
        value = item["value"]

        canonical = FIELD_MAP.get(key, key)

        result[canonical] = value.strip()

    return result
```

```python
LEFT_ANCHORS = [
    "INVOICE NO",
    "B/L NO",
    "거래처",
    "사업자번호",
    "담당자",
    "SHIPPER",
    "VESSEL/VOY",
    "P.O.L.",
    "P.O.D.",
    "WEIGHT",
    "MEASUREMENT"
]

RIGHT_ANCHORS = [
    "B/L TYPE",
    "출력일자",
    "MRN",
    "MSN/HSN",
    "Page",
    "ONBOARD/ARRIVAL",
    "PKG'S",
    "TERM"
]
def reconstruct_form(doc, group_ref):

    items = extract_text_items(doc, group_ref)

    rows = cluster_rows(items)

    left_rows, right_rows = split_columns(rows)

    left = parse_rows(left_rows, LEFT_ANCHORS)
    right = parse_rows(right_rows, RIGHT_ANCHORS)

    result = {}

    result.update(left)
    result.update(right)

    return result

reconstruct_form(doc_dict, "groups/1")
```

```python
def resolve_ref(doc: dict, ref: str) -> dict:
    if ref.startswith("#/"):
        ref = ref[2:]
    elif ref.startswith("/"):
        ref = ref[1:]

    cur = doc
    for p in ref.split("/"):
        cur = cur[int(p)] if p.isdigit() else cur[p]
    return cur


def get_first_prov(obj: dict):
    prov = obj.get("prov") or []
    return prov[0] if prov else None


def get_group_bbox(group_obj: dict, doc: dict):
    boxes = []

    for child in group_obj.get("children", []):
        ref = child.get("$ref")
        if not ref:
            continue

        item = resolve_ref(doc, ref)
        prov = get_first_prov(item)

        if prov and "bbox" in prov:
            boxes.append(prov["bbox"])

    if not boxes:
        return None

    coord_origin = boxes[0].get("coord_origin", "BOTTOMLEFT")

    return {
        "l": min(b.get("l", 999999) for b in boxes),
        "t": max(b.get("t", -999999) for b in boxes),  # BOTTOMLEFT 기준 위쪽
        "r": max(b.get("r", -999999) for b in boxes),
        "b": min(b.get("b", 999999) for b in boxes),
        "coord_origin": coord_origin,
    }


def get_obj_bbox(obj: dict, doc: dict):
    """
    일반 객체면 prov bbox 사용
    group이면 children bbox를 합쳐서 대표 bbox 생성
    """
    # group 판별: self_ref가 groups/... 이거나 name == group
    self_ref = obj.get("self_ref", "")
    name = obj.get("name")

    if self_ref.startswith("#/groups/") or name == "group":
        return get_group_bbox(obj, doc)

    prov = get_first_prov(obj)
    if prov and "bbox" in prov:
        return prov["bbox"]

    return None


def get_bbox_sort_key(obj: dict, doc: dict):
    bbox = get_obj_bbox(obj, doc)

    if not bbox:
        return (9999, 999999, 999999)

    # page_no는 일반 객체는 prov에서, group은 child page_no 중 첫번째/최소값 사용
    page_no = 9999

    self_ref = obj.get("self_ref", "")
    name = obj.get("name")

    if self_ref.startswith("#/groups/") or name == "group":
        page_nos = []
        for child in obj.get("children", []):
            ref = child.get("$ref")
            if not ref:
                continue
            item = resolve_ref(doc, ref)
            prov = get_first_prov(item)
            if prov:
                page_nos.append(prov.get("page_no", 9999))
        if page_nos:
            page_no = min(page_nos)
    else:
        prov = get_first_prov(obj)
        if prov:
            page_no = prov.get("page_no", 9999)

    coord_origin = bbox.get("coord_origin", "BOTTOMLEFT")
    l = bbox.get("l", 999999)
    t = bbox.get("t", 999999)

    # BOTTOMLEFT면 t가 클수록 화면 위쪽
    if coord_origin == "BOTTOMLEFT":
        return (page_no, -t, l)

    # TOPLEFT면 t가 작을수록 화면 위쪽
    return (page_no, t, l)


def get_label_and_text_from_obj(obj: dict, doc: dict):
    label = obj.get("label")
    text = obj.get("text", "")

    # group은 직접 text가 없으니 children text를 합쳐서 preview 생성
    if not text:
        self_ref = obj.get("self_ref", "")
        name = obj.get("name")

        if self_ref.startswith("#/groups/") or name == "group":
            parts = []
            for child in obj.get("children", []):
                ref = child.get("$ref")
                if not ref:
                    continue
                item = resolve_ref(doc, ref)
                child_text = item.get("text", "")
                if child_text:
                    parts.append(child_text)
            text = " ".join(parts)

    return label, text


body = doc_dict.get("body") or {}
children = body.get("children") or []

blocks = []
for child in children:
    ref = child.get("$ref")
    if not ref:
        continue

    obj = resolve_ref(doc_dict, ref)
    blocks.append((ref, obj))

blocks_sorted = sorted(blocks, key=lambda x: get_bbox_sort_key(x[1], doc_dict))

print("bbox 기준 시각 순서:")
for i, (ref, obj) in enumerate(blocks_sorted):
    label, text = get_label_and_text_from_obj(obj, doc_dict)
    bbox = get_obj_bbox(obj, doc_dict)

    print(
        f"{i}: {ref}  " f"label={label}  " f"text={(text or '')[:80]}  " f"bbox={bbox}"
    )
    print()
```

```python
import re
from typing import List, Dict, Tuple, Optional


def resolve_ref(doc: dict, ref: str) -> dict:
    if ref.startswith("#/"):
        ref = ref[2:]
    elif ref.startswith("/"):
        ref = ref[1:]

    cur = doc
    for p in ref.split("/"):
        cur = cur[int(p)] if p.isdigit() else cur[p]
    return cur


def get_first_prov(obj: dict):
    prov = obj.get("prov") or []
    return prov[0] if prov else None


def extract_group_items(doc: dict, group_ref: str) -> List[dict]:
    """
    group children에서 text + bbox를 추출
    """
    group_obj = resolve_ref(doc, group_ref)
    items = []

    for child in group_obj.get("children", []):
        ref = child.get("$ref")
        if not ref:
            continue

        obj = resolve_ref(doc, ref)
        text = (obj.get("text") or "").strip()
        prov = get_first_prov(obj)

        if not text or not prov or "bbox" not in prov:
            continue

        items.append(
            {
                "ref": ref,
                "text": text,
                "bbox": prov["bbox"],
                "page_no": prov.get("page_no", 9999),
                "label": obj.get("label"),
            }
        )

    return items


def get_sort_key_for_item(item: dict):
    bbox = item["bbox"]
    page_no = item.get("page_no", 9999)
    coord_origin = bbox.get("coord_origin", "BOTTOMLEFT")
    l = bbox.get("l", 999999)
    t = bbox.get("t", 999999)

    if coord_origin == "BOTTOMLEFT":
        return (page_no, -t, l)
    return (page_no, t, l)


def cluster_rows(items: List[dict], y_tol: float = 10.0) -> List[dict]:
    """
    같은 y축 근처 항목들을 한 row로 묶음
    """
    if not items:
        return []

    items_sorted = sorted(items, key=get_sort_key_for_item)
    rows = []

    for item in items_sorted:
        bbox = item["bbox"]
        coord_origin = bbox.get("coord_origin", "BOTTOMLEFT")
        y = bbox.get("t", 0)

        placed = False
        for row in rows:
            if row["coord_origin"] != coord_origin:
                continue

            if abs(row["y"] - y) <= y_tol:
                row["items"].append(item)
                row["ys"].append(y)
                row["y"] = sum(row["ys"]) / len(row["ys"])
                placed = True
                break

        if not placed:
            rows.append(
                {
                    "coord_origin": coord_origin,
                    "y": y,
                    "ys": [y],
                    "items": [item],
                }
            )

    for row in rows:
        row["items"] = sorted(row["items"], key=lambda x: x["bbox"].get("l", 999999))

    rows = sorted(
        rows,
        key=lambda r: (
            r["items"][0].get("page_no", 9999),
            -r["y"] if r["coord_origin"] == "BOTTOMLEFT" else r["y"],
            min(i["bbox"].get("l", 999999) for i in r["items"]),
        ),
    )
    return rows


def zone_to_text(zone_items: List[dict]) -> str:
    parts = []
    for item in zone_items:
        t = (item.get("text") or "").strip()
        if t:
            parts.append(t)

    text = " ".join(parts)
    text = re.sub(r"\s*:\s*", " : ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


def split_row_into_zones_adaptive(
    row_items: List[dict], min_threshold: float = 18.0, ratio: float = 1.8
) -> List[List[dict]]:
    """
    row 내부 인접 item 간 x-gap을 보고 adaptive threshold로 zone 분리
    row마다 2존/3존/그 이상도 가능
    """
    if not row_items:
        return []

    items = sorted(row_items, key=lambda x: x["bbox"].get("l", 999999))

    if len(items) == 1:
        return [items]

    gaps = []
    for prev, curr in zip(items, items[1:]):
        prev_r = prev["bbox"].get("r", 0)
        curr_l = curr["bbox"].get("l", 0)
        gap = max(0, curr_l - prev_r)
        gaps.append(gap)

    avg_gap = sum(gaps) / len(gaps) if gaps else 0
    threshold = max(min_threshold, avg_gap * ratio)

    zones = []
    current_zone = [items[0]]

    for i, curr in enumerate(items[1:]):
        prev = items[i]
        gap = max(0, curr["bbox"].get("l", 0) - prev["bbox"].get("r", 0))

        if gap >= threshold:
            zones.append(current_zone)
            current_zone = [curr]
        else:
            current_zone.append(curr)

    zones.append(current_zone)
    return zones


def normalize_key(key: str) -> str:
    key = key.strip()
    key = re.sub(r"\s+", " ", key)
    key = key.replace(" :", "").strip()
    return key


def parse_zone_kv(zone_items: List[dict]) -> Tuple[Optional[str], Optional[str]]:
    """
    zone 하나를 key-value로 해석
    """
    text = zone_to_text(zone_items)
    if not text:
        return None, None

    # 1) 가장 일반적: key : value
    if " : " in text:
        key, value = text.split(" : ", 1)
        return normalize_key(key), value.strip()

    # 2) ':'가 붙어있는 경우
    if ":" in text:
        key, value = text.split(":", 1)
        return normalize_key(key), value.strip()

    # 3) 콜론이 따로 박스여서 key/value가 분리된 경우 보정
    # 예: ["Weight", ":", "134,578.000 KGS"]
    tokens = [i["text"].strip() for i in zone_items if i.get("text", "").strip()]
    if ":" in tokens:
        idx = tokens.index(":")
        key = " ".join(tokens[:idx]).strip()
        value = " ".join(tokens[idx + 1 :]).strip()
        if key:
            return normalize_key(key), value

    # 4) fallback: 첫 토큰을 key로 보기엔 위험하므로 전체를 key로 반환
    return normalize_key(text), ""


def merge_multiline_kv(
    rows_zone_kv: List[List[Tuple[Optional[str], Optional[str]]]],
    known_keys: Optional[set] = None,
) -> Dict[str, str]:
    """
    row별 zone parsing 결과를 최종 dict로 합침.
    새 key가 없고 value만 이어지는 줄은 이전 key에 붙이는 식으로 보정 가능.
    """
    result: Dict[str, str] = {}
    last_key: Optional[str] = None

    for row_zones in rows_zone_kv:
        for key, value in row_zones:
            if not key:
                if last_key and value:
                    result[last_key] = (result[last_key] + " " + value).strip()
                continue

            key_norm = normalize_key(key)

            # known_keys가 있고 현재 key가 unknown이면 continuation처럼 처리할 수도 있음
            if known_keys is not None and key_norm not in known_keys and last_key:
                append_text = " ".join([x for x in [key_norm, value] if x]).strip()
                result[last_key] = (result[last_key] + " " + append_text).strip()
                continue

            if key_norm in result and value:
                # 동일 key 재등장 시 이어붙임
                result[key_norm] = (result[key_norm] + " " + value).strip()
            else:
                result[key_norm] = value.strip() if value else ""

            last_key = key_norm

    return result


def print_row_zone_debug(
    rows: List[dict], zones_by_row: List[List[List[dict]]]
) -> None:
    for row_idx, (row, zones) in enumerate(zip(rows, zones_by_row)):
        row_text = zone_to_text(row["items"])
        print(f"\nROW[{row_idx}] {row_text}")
        for zi, zone in enumerate(zones):
            print(f"  zone[{zi}] = {zone_to_text(zone)}")


def parse_group_kv_with_adaptive_zones(
    doc: dict,
    group_ref: str,
    y_tol: float = 10.0,
    min_gap_threshold: float = 18.0,
    gap_ratio: float = 1.8,
    known_keys: Optional[set] = None,
    debug: bool = True,
) -> Dict[str, str]:
    """
    통합 실행 함수
    """
    items = extract_group_items(doc, group_ref)
    rows = cluster_rows(items, y_tol=y_tol)

    zones_by_row: List[List[List[dict]]] = []
    kv_by_row: List[List[Tuple[Optional[str], Optional[str]]]] = []

    for row in rows:
        zones = split_row_into_zones_adaptive(
            row["items"], min_threshold=min_gap_threshold, ratio=gap_ratio
        )
        zones_by_row.append(zones)

        row_kvs = []
        for zone in zones:
            key, value = parse_zone_kv(zone)
            row_kvs.append((key, value))
        kv_by_row.append(row_kvs)

    if debug:
        print_row_zone_debug(rows, zones_by_row)
        print("\nPARSED KV BY ROW:")
        for i, row_kvs in enumerate(kv_by_row):
            print(f"ROW[{i}] -> {row_kvs}")

    result = merge_multiline_kv(kv_by_row, known_keys=known_keys)
    return result
```

```python
import re
from typing import List, Dict, Tuple, Optional


# =========================
# 1. 기본 유틸
# =========================


def resolve_ref(doc: dict, ref: str) -> dict:
    if ref.startswith("#/"):
        ref = ref[2:]
    elif ref.startswith("/"):
        ref = ref[1:]

    cur = doc
    for p in ref.split("/"):
        cur = cur[int(p)] if p.isdigit() else cur[p]
    return cur


def get_first_prov(obj: dict):
    prov = obj.get("prov") or []
    return prov[0] if prov else None


def normalize_text(text: str) -> str:
    text = (text or "").strip()
    text = re.sub(r"\s+", " ", text)
    return text


def normalize_key(key: str) -> str:
    key = normalize_text(key)
    key = key.replace(" :", "").strip()
    return key


# =========================
# 2. group 원본 출력용
# =========================


def get_group_texts_in_order(doc: dict, group_obj: dict) -> List[str]:
    items = []

    for child in group_obj.get("children", []):
        ref = child.get("$ref")
        if not ref:
            continue

        item = resolve_ref(doc, ref)
        text = normalize_text(item.get("text", ""))
        prov = item.get("prov") or []

        if prov and text:
            bbox = prov[0].get("bbox", {})
            coord_origin = bbox.get("coord_origin", "BOTTOMLEFT")
            l = bbox.get("l", 999999)
            t = bbox.get("t", 999999)

            if coord_origin == "BOTTOMLEFT":
                sort_key = (-t, l)
            else:
                sort_key = (t, l)

            items.append((sort_key, text, ref))

    items.sort(key=lambda x: x[0])
    return [text for _, text, _ in items]


# =========================
# 3. table 출력용
# =========================


def print_table_rows(table_obj: dict):
    grid = table_obj.get("data", {}).get("grid", [])

    for row_idx, row in enumerate(grid):
        row_values = [(cell.get("text") or "").strip() for cell in row]
        print(f"    row[{row_idx}] = {row_values}")


# =========================
# 4. group children -> text+bbox item 추출
# =========================


def extract_group_items(doc: dict, group_ref: str) -> List[dict]:
    group_obj = resolve_ref(doc, group_ref)
    items = []

    for child in group_obj.get("children", []):
        ref = child.get("$ref")
        if not ref:
            continue

        obj = resolve_ref(doc, ref)
        text = normalize_text(obj.get("text", ""))
        prov = get_first_prov(obj)

        if not text or not prov or "bbox" not in prov:
            continue

        items.append(
            {
                "ref": ref,
                "text": text,
                "bbox": prov["bbox"],
                "page_no": prov.get("page_no", 9999),
                "label": obj.get("label"),
            }
        )

    return items


# =========================
# 5. row clustering
# =========================


def get_sort_key_for_item(item: dict):
    bbox = item["bbox"]
    page_no = item.get("page_no", 9999)
    coord_origin = bbox.get("coord_origin", "BOTTOMLEFT")
    l = bbox.get("l", 999999)
    t = bbox.get("t", 999999)

    if coord_origin == "BOTTOMLEFT":
        return (page_no, -t, l)
    return (page_no, t, l)


def cluster_rows(items: List[dict], y_tol: float = 10.0) -> List[dict]:
    if not items:
        return []

    items_sorted = sorted(items, key=get_sort_key_for_item)
    rows = []

    for item in items_sorted:
        bbox = item["bbox"]
        coord_origin = bbox.get("coord_origin", "BOTTOMLEFT")
        y = bbox.get("t", 0)

        placed = False
        for row in rows:
            if row["coord_origin"] != coord_origin:
                continue

            if abs(row["y"] - y) <= y_tol:
                row["items"].append(item)
                row["ys"].append(y)
                row["y"] = sum(row["ys"]) / len(row["ys"])
                placed = True
                break

        if not placed:
            rows.append(
                {
                    "coord_origin": coord_origin,
                    "y": y,
                    "ys": [y],
                    "items": [item],
                }
            )

    for row in rows:
        row["items"] = sorted(row["items"], key=lambda x: x["bbox"].get("l", 999999))

    rows = sorted(
        rows,
        key=lambda r: (
            r["items"][0].get("page_no", 9999),
            -r["y"] if r["coord_origin"] == "BOTTOMLEFT" else r["y"],
            min(i["bbox"].get("l", 999999) for i in r["items"]),
        ),
    )
    return rows


# =========================
# 6. adaptive zone split
# =========================


def zone_to_text(zone_items: List[dict]) -> str:
    parts = []
    for item in zone_items:
        t = normalize_text(item.get("text", ""))
        if t:
            parts.append(t)

    text = " ".join(parts)
    text = re.sub(r"\s*:\s*", " : ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


def split_row_into_zones_adaptive(
    row_items: List[dict], min_threshold: float = 18.0, ratio: float = 1.8
) -> List[List[dict]]:
    if not row_items:
        return []

    items = sorted(row_items, key=lambda x: x["bbox"].get("l", 999999))

    if len(items) == 1:
        return [items]

    gaps = []
    for prev, curr in zip(items, items[1:]):
        prev_r = prev["bbox"].get("r", 0)
        curr_l = curr["bbox"].get("l", 0)
        gap = max(0, curr_l - prev_r)
        gaps.append(gap)

    avg_gap = sum(gaps) / len(gaps) if gaps else 0
    threshold = max(min_threshold, avg_gap * ratio)

    zones = []
    current_zone = [items[0]]

    for i, curr in enumerate(items[1:]):
        prev = items[i]
        gap = max(0, curr["bbox"].get("l", 0) - prev["bbox"].get("r", 0))

        if gap >= threshold:
            zones.append(current_zone)
            current_zone = [curr]
        else:
            current_zone.append(curr)

    zones.append(current_zone)
    return zones


# =========================
# 7. zone -> key/value
# =========================


def parse_zone_kv(zone_items: List[dict]) -> Tuple[Optional[str], Optional[str]]:
    text = zone_to_text(zone_items)
    if not text:
        return None, None

    if " : " in text:
        key, value = text.split(" : ", 1)
        return normalize_key(key), normalize_text(value)

    if ":" in text:
        key, value = text.split(":", 1)
        return normalize_key(key), normalize_text(value)

    tokens = [
        normalize_text(i["text"])
        for i in zone_items
        if normalize_text(i.get("text", ""))
    ]
    if ":" in tokens:
        idx = tokens.index(":")
        key = " ".join(tokens[:idx]).strip()
        value = " ".join(tokens[idx + 1 :]).strip()
        if key:
            return normalize_key(key), normalize_text(value)

    return normalize_key(text), ""


def merge_multiline_kv(
    rows_zone_kv: List[List[Tuple[Optional[str], Optional[str]]]],
    known_keys: Optional[set] = None,
) -> Dict[str, str]:
    result: Dict[str, str] = {}
    last_key: Optional[str] = None

    for row_zones in rows_zone_kv:
        for key, value in row_zones:
            if not key:
                if last_key and value:
                    result[last_key] = normalize_text(result[last_key] + " " + value)
                continue

            key_norm = normalize_key(key)

            if known_keys is not None and key_norm not in known_keys and last_key:
                append_text = normalize_text(
                    " ".join([x for x in [key_norm, value] if x])
                )
                result[last_key] = normalize_text(result[last_key] + " " + append_text)
                continue

            if key_norm in result and value:
                result[key_norm] = normalize_text(result[key_norm] + " " + value)
            else:
                result[key_norm] = normalize_text(value) if value else ""

            last_key = key_norm

    return result


# =========================
# 8. group pretty parse
# =========================


def parse_group_kv_with_adaptive_zones(
    doc: dict,
    group_ref: str,
    y_tol: float = 10.0,
    min_gap_threshold: float = 18.0,
    gap_ratio: float = 1.8,
    known_keys: Optional[set] = None,
) -> Dict[str, str]:
    items = extract_group_items(doc, group_ref)
    rows = cluster_rows(items, y_tol=y_tol)

    kv_by_row: List[List[Tuple[Optional[str], Optional[str]]]] = []

    for row in rows:
        zones = split_row_into_zones_adaptive(
            row["items"], min_threshold=min_gap_threshold, ratio=gap_ratio
        )

        row_kvs = []
        for zone in zones:
            key, value = parse_zone_kv(zone)
            row_kvs.append((key, value))
        kv_by_row.append(row_kvs)

    result = merge_multiline_kv(kv_by_row, known_keys=known_keys)
    return result


def print_group_pretty(
    doc: dict,
    ref: str,
    obj: dict,
    group_known_keys_map: Optional[Dict[str, set]] = None,
    y_tol: float = 10.0,
    min_gap_threshold: float = 18.0,
    gap_ratio: float = 1.8,
):
    known_keys = None
    if group_known_keys_map:
        known_keys = group_known_keys_map.get(ref)

    try:
        parsed = parse_group_kv_with_adaptive_zones(
            doc=doc,
            group_ref=ref,
            y_tol=y_tol,
            min_gap_threshold=min_gap_threshold,
            gap_ratio=gap_ratio,
            known_keys=known_keys,
        )

        # parse 결과가 있으면 key:value로 출력
        if parsed:
            for k, v in parsed.items():
                print(f"    - {k}: {v}")
            return

    except Exception:
        pass

    # fallback: 원래 child 순서 출력
    texts = get_group_texts_in_order(doc, obj)
    for t in texts:
        print(f"    - {t}")


# =========================
# 9. 전체 block 출력
# =========================


def print_blocks_content_in_order(
    doc: dict,
    blocks_sorted,
    group_known_keys_map: Optional[Dict[str, set]] = None,
    y_tol: float = 10.0,
    min_gap_threshold: float = 18.0,
    gap_ratio: float = 1.8,
):
    for i, (ref, obj) in enumerate(blocks_sorted):
        label = obj.get("label")
        self_ref = obj.get("self_ref", "")
        name = obj.get("name", "")

        print(f"\n[{i}] {ref}  label={label}")

        # text
        if "text" in obj and obj.get("text"):
            print(obj["text"])
            continue

        # group
        if self_ref.startswith("#/groups/") or name == "group":
            print_group_pretty(
                doc=doc,
                ref=ref,
                obj=obj,
                group_known_keys_map=group_known_keys_map,
                y_tol=y_tol,
                min_gap_threshold=min_gap_threshold,
                gap_ratio=gap_ratio,
            )
            continue

        # table
        if self_ref.startswith("#/tables/"):
            print_table_rows(obj)
            continue

        # picture
        if self_ref.startswith("#/pictures/") or label == "picture":
            print("    [PICTURE]")
            continue

        print("    [NO RENDERABLE CONTENT]")
```

```python
# document_parsing_engine/app/services/layout_segment_mapping_service.py

from __future__ import annotations

from dataclasses import dataclass, field
from enum import Enum
from typing import Any, Protocol


# =========================================================
# Input Model
# =========================================================

@dataclass
class BlockContent:
    ref: str
    label: str | None
    content_type: str
    content: Any


# =========================================================
# Segment Enums / Definitions
# =========================================================

class SegmentType(str, Enum):
    QUOTATION_META = "quotation_meta"
    ITEMS = "items"
    REMARKS = "remarks"


@dataclass(frozen=True)
class SegmentDefinition:
    doc_type: str
    allowed_segments: tuple[SegmentType, ...]


class SegmentDefinitions:
    """
    doc_type 별 허용 segment 집합
    """

    QUOTATION = SegmentDefinition(
        doc_type="quotation",
        allowed_segments=(
            SegmentType.QUOTATION_META,
            SegmentType.ITEMS,
            SegmentType.REMARKS,
        ),
    )

    @classmethod
    def get_by_doc_type(cls, doc_type: str) -> SegmentDefinition:
        mapping = {
            "quotation": cls.QUOTATION,
            "steel_quotation": cls.QUOTATION,
        }
        return mapping.get(doc_type, cls.QUOTATION)


# =========================================================
# Output Models
# =========================================================

@dataclass
class SegmentCandidate:
    segment_type: SegmentType
    score: float
    reasons: list[str] = field(default_factory=list)


@dataclass
class BlockSegmentRecommendation:
    block_ref: str
    label: str | None
    content_type: str
    primary_segment: SegmentType | None
    candidates: list[SegmentCandidate] = field(default_factory=list)


@dataclass
class SegmentBucket:
    segment_type: SegmentType
    block_refs: list[str] = field(default_factory=list)
    reasons: list[str] = field(default_factory=list)


@dataclass
class LayoutSegmentMappingRecommendation:
    doc_type: str
    allowed_segments: list[str]
    block_recommendations: list[BlockSegmentRecommendation] = field(default_factory=list)
    segment_buckets: list[SegmentBucket] = field(default_factory=list)
    unmapped_block_refs: list[str] = field(default_factory=list)
    warnings: list[str] = field(default_factory=list)


# =========================================================
# Rule Interfaces
# =========================================================

class BlockSegmentRule(Protocol):
    """
    하나의 block에 대해 segment 후보 점수를 계산하는 규칙
    """

    def supports(
        self,
        *,
        doc_type: str,
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
    ) -> bool:
        ...

    def score(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        block: BlockContent,
        allowed_segments: tuple[SegmentType, ...],
    ) -> list[SegmentCandidate]:
        ...


# =========================================================
# Base Utility
# =========================================================

class BaseBlockSegmentRule:
    def supports(
        self,
        *,
        doc_type: str,
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
    ) -> bool:
        return True

    @staticmethod
    def _normalize_text(text: str) -> str:
        return " ".join(text.strip().lower().split())

    @classmethod
    def _text_of(cls, block: BlockContent) -> str:
        if block.content is None:
            return ""

        if isinstance(block.content, str):
            return block.content.strip()

        if isinstance(block.content, dict):
            parts: list[str] = []
            for k, v in block.content.items():
                key = "" if k is None else str(k).strip()
                val = "" if v is None else str(v).strip()
                if key:
                    parts.append(key)
                if val:
                    parts.append(val)
            return " ".join(parts).strip()

        if isinstance(block.content, list):
            parts: list[str] = []
            for row in block.content:
                if isinstance(row, list):
                    for cell in row:
                        cell_text = "" if cell is None else str(cell).strip()
                        if cell_text:
                            parts.append(cell_text)
                else:
                    row_text = "" if row is None else str(row).strip()
                    if row_text:
                        parts.append(row_text)
            return " ".join(parts).strip()

        return str(block.content).strip()

    @classmethod
    def _contains_any(cls, text: str, keywords: list[str]) -> bool:
        normalized = cls._normalize_text(text)
        return any(cls._normalize_text(keyword) in normalized for keyword in keywords)

    @classmethod
    def _count_matches(cls, text: str, keywords: list[str]) -> int:
        normalized = cls._normalize_text(text)
        return sum(1 for keyword in keywords if cls._normalize_text(keyword) in normalized)

    @staticmethod
    def _candidate(
        segment_type: SegmentType,
        score: float,
        *reasons: str,
    ) -> SegmentCandidate:
        return SegmentCandidate(
            segment_type=segment_type,
            score=max(0.0, min(score, 0.99)),
            reasons=[r for r in reasons if r],
        )


# =========================================================
# Concrete Rules
# =========================================================

class TableItemsRule(BaseBlockSegmentRule):
    """
    table 계열이면 items 우선 추천
    """

    TABLE_ITEM_KEYWORDS = [
        "품명", "재질", "규격", "길이", "수량", "중량", "단가", "금액", "비고",
        "length", "qty", "q'ty", "weight", "unit price", "amount", "remark", "remarks",
        "maker", "unit", "합계", "부가세",
    ]

    def supports(
        self,
        *,
        doc_type: str,
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
    ) -> bool:
        return block.content_type == "table"

    def score(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        block: BlockContent,
        allowed_segments: tuple[SegmentType, ...],
    ) -> list[SegmentCandidate]:
        text = self._text_of(block)

        score = 0.72
        reasons = ["content_type=table"]

        match_count = self._count_matches(text, self.TABLE_ITEM_KEYWORDS)
        if match_count >= 3:
            score += 0.18
            reasons.append(f"item-table keywords matched ({match_count})")
        elif match_count >= 1:
            score += 0.10
            reasons.append(f"table keywords matched ({match_count})")

        return [self._candidate(SegmentType.ITEMS, score, *reasons)]


class KeyValueMetaRule(BaseBlockSegmentRule):
    """
    key-value area 중 메타/거래처/거래조건 후보
    """

    META_KEYWORDS = [
        "견적 일자", "견적일자", "견적번호", "no.", "프로젝트", "의뢰번호",
        "수 신", "참 조", "작성자", "대표 이사", "대표이사",
        "유효기간", "지불조건", "결제조건", "인도조건", "납기일자",
        "tel", "fax", "e-mail", "email", "페이지",
        "vat포함", "합계금액",
    ]

    def supports(
        self,
        *,
        doc_type: str,
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
    ) -> bool:
        return block.content_type in {"group_kv", "text"}

    def score(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        block: BlockContent,
        allowed_segments: tuple[SegmentType, ...],
    ) -> list[SegmentCandidate]:
        text = self._text_of(block)
        if not text:
            return []

        match_count = self._count_matches(text, self.META_KEYWORDS)
        if match_count == 0:
            return []

        score = 0.58
        reasons = [f"meta keywords matched ({match_count})"]

        if block.content_type == "group_kv":
            score += 0.14
            reasons.append("content_type=group_kv")

        if block.label == "section_header":
            score += 0.05
            reasons.append("label=section_header")

        if "견적서" in text:
            score += 0.08
            reasons.append("quotation title matched")

        return [self._candidate(SegmentType.QUOTATION_META, score, *reasons)]


class QuotationTitleRule(BaseBlockSegmentRule):
    def supports(
        self,
        *,
        doc_type: str,
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
    ) -> bool:
        return block.content_type == "text"

    def score(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        block: BlockContent,
        allowed_segments: tuple[SegmentType, ...],
    ) -> list[SegmentCandidate]:
        text = self._text_of(block)
        if not text:
            return []

        normalized = self._normalize_text(text)
        if normalized in {"견적서", "quotation", "quote"}:
            return [
                self._candidate(
                    SegmentType.QUOTATION_META,
                    0.80,
                    "document title matched",
                )
            ]

        if "견적서" in normalized:
            return [
                self._candidate(
                    SegmentType.QUOTATION_META,
                    0.68,
                    "quotation title contained",
                )
            ]

        return []


class HeaderMetaFallbackRule(BaseBlockSegmentRule):
    """
    section_header/text 계열에서 상단 메타성 텍스트 fallback
    """

    HEADER_META_KEYWORDS = [
        "귀중", "貴中", "견적 일자", "견적일자", "no.", "tel", "fax", "email", "e-mail"
    ]

    def supports(
        self,
        *,
        doc_type: str,
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
    ) -> bool:
        return block.content_type == "text"

    def score(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        block: BlockContent,
        allowed_segments: tuple[SegmentType, ...],
    ) -> list[SegmentCandidate]:
        text = self._text_of(block)
        if not text:
            return []

        match_count = self._count_matches(text, self.HEADER_META_KEYWORDS)
        if match_count == 0:
            return []

        score = 0.54
        reasons = [f"header meta keywords matched ({match_count})"]

        if block.label == "section_header":
            score += 0.06
            reasons.append("label=section_header")

        return [self._candidate(SegmentType.QUOTATION_META, score, *reasons)]


class RemarksKeywordRule(BaseBlockSegmentRule):
    """
    비고 / 특기사항 / 자유문장
    """

    REMARK_KEYWORDS = [
        "특기사항", "remarks", "remark", "비고", "note", "기타", "특이사항",
        "대체견적", "생산불가", "수급불가", "로 견적 합니다", "로 견적합니다",
    ]

    def score(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        block: BlockContent,
        allowed_segments: tuple[SegmentType, ...],
    ) -> list[SegmentCandidate]:
        text = self._text_of(block)
        if not text:
            return []

        match_count = self._count_matches(text, self.REMARK_KEYWORDS)
        if match_count == 0:
            return []

        score = 0.66
        reasons = [f"remarks keywords matched ({match_count})"]

        if block.content_type == "group_kv":
            score += 0.05
            reasons.append("content_type=group_kv")

        if block.label == "list":
            score += 0.08
            reasons.append("label=list")

        if "특기사항" in text or "비고" in text:
            score += 0.08
            reasons.append("remarks title matched")

        return [self._candidate(SegmentType.REMARKS, score, *reasons)]


class PageFooterIgnoreRule(BaseBlockSegmentRule):
    """
    footer는 기본적으로 segment 미분류
    """

    def supports(
        self,
        *,
        doc_type: str,
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
    ) -> bool:
        return block.label == "page_footer"

    def score(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        block: BlockContent,
        allowed_segments: tuple[SegmentType, ...],
    ) -> list[SegmentCandidate]:
        return []


class PictureIgnoreRule(BaseBlockSegmentRule):
    """
    picture는 기본적으로 미분류
    """

    def supports(
        self,
        *,
        doc_type: str,
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
    ) -> bool:
        return block.content_type == "picture"

    def score(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        block: BlockContent,
        allowed_segments: tuple[SegmentType, ...],
    ) -> list[SegmentCandidate]:
        return []


# =========================================================
# Rule Registry
# =========================================================

class SegmentRuleRegistry:
    """
    doc_type 별 규칙 세트 제공
    """

    def get_rules(self, doc_type: str) -> list[BlockSegmentRule]:
        quotation_rules: list[BlockSegmentRule] = [
            PageFooterIgnoreRule(),
            PictureIgnoreRule(),
            TableItemsRule(),
            RemarksKeywordRule(),
            QuotationTitleRule(),
            KeyValueMetaRule(),
            HeaderMetaFallbackRule(),
        ]

        mapping = {
            "quotation": quotation_rules,
            "steel_quotation": quotation_rules,
        }

        return mapping.get(doc_type, quotation_rules)


# =========================================================
# Service
# =========================================================

class LayoutSegmentMappingService:
    """
    block 단위 입력을 받아 semantic segment 추천을 수행한다.

    현재 quotation 기준 segment:
    - quotation_meta
    - items
    - remarks
    """

    def __init__(
        self,
        rule_registry: SegmentRuleRegistry | None = None,
        primary_threshold: float = 0.50,
    ) -> None:
        self._rule_registry = rule_registry or SegmentRuleRegistry()
        self._primary_threshold = primary_threshold

    def recommend(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        blocks: list[BlockContent],
    ) -> LayoutSegmentMappingRecommendation:
        segment_def = SegmentDefinitions.get_by_doc_type(doc_type)
        rules = self._rule_registry.get_rules(doc_type)

        normalized_blocks = self._normalize_blocks(blocks)

        block_recommendations = [
            self._recommend_for_block(
                doc_type=doc_type,
                doc_dict=doc_dict,
                allowed_segments=segment_def.allowed_segments,
                block=block,
                rules=rules,
            )
            for block in normalized_blocks
        ]

        segment_buckets = self._build_segment_buckets(
            allowed_segments=segment_def.allowed_segments,
            block_recommendations=block_recommendations,
        )
        unmapped_block_refs = self._collect_unmapped_block_refs(block_recommendations)
        warnings = self._build_warnings(
            doc_type=doc_type,
            block_recommendations=block_recommendations,
            unmapped_block_refs=unmapped_block_refs,
        )

        return LayoutSegmentMappingRecommendation(
            doc_type=doc_type,
            allowed_segments=[segment.value for segment in segment_def.allowed_segments],
            block_recommendations=block_recommendations,
            segment_buckets=segment_buckets,
            unmapped_block_refs=unmapped_block_refs,
            warnings=warnings,
        )

    # -----------------------------------------------------
    # Internal Steps
    # -----------------------------------------------------

    def _normalize_blocks(
        self,
        blocks: list[BlockContent],
    ) -> list[BlockContent]:
        """
        content 정규화
        """
        normalized: list[BlockContent] = []

        for block in blocks:
            content = block.content

            if isinstance(content, str):
                content = content.strip()

            elif isinstance(content, dict):
                normalized_dict: dict[str, Any] = {}
                for k, v in content.items():
                    nk = "" if k is None else str(k).strip()
                    nv = "" if v is None else str(v).strip()
                    normalized_dict[nk] = nv
                content = normalized_dict

            elif isinstance(content, list):
                normalized_list: list[Any] = []
                for row in content:
                    if isinstance(row, list):
                        normalized_row = [
                            "" if cell is None else str(cell).strip()
                            for cell in row
                        ]
                        normalized_list.append(normalized_row)
                    else:
                        normalized_list.append("" if row is None else str(row).strip())
                content = normalized_list

            normalized.append(
                BlockContent(
                    ref=block.ref.strip(),
                    label=block.label.strip() if isinstance(block.label, str) else block.label,
                    content_type=block.content_type.strip(),
                    content=content,
                )
            )

        return normalized

    def _recommend_for_block(
        self,
        *,
        doc_type: str,
        doc_dict: dict[str, Any],
        allowed_segments: tuple[SegmentType, ...],
        block: BlockContent,
        rules: list[BlockSegmentRule],
    ) -> BlockSegmentRecommendation:
        raw_candidates: list[SegmentCandidate] = []

        for rule in rules:
            if not rule.supports(
                doc_type=doc_type,
                allowed_segments=allowed_segments,
                block=block,
            ):
                continue

            candidates = rule.score(
                doc_type=doc_type,
                doc_dict=doc_dict,
                block=block,
                allowed_segments=allowed_segments,
            )
            raw_candidates.extend(
                candidate
                for candidate in candidates
                if candidate.segment_type in allowed_segments
            )

        merged_candidates = self._merge_candidates(raw_candidates)
        primary_segment = self._select_primary_segment(merged_candidates)

        return BlockSegmentRecommendation(
            block_ref=block.ref,
            label=block.label,
            content_type=block.content_type,
            primary_segment=primary_segment,
            candidates=merged_candidates,
        )

    def _merge_candidates(
        self,
        candidates: list[SegmentCandidate],
    ) -> list[SegmentCandidate]:
        """
        동일 segment_type 후보를 병합
        - score: max 사용
        - reasons: dedupe
        """
        merged: dict[SegmentType, SegmentCandidate] = {}

        for candidate in candidates:
            existing = merged.get(candidate.segment_type)
            if existing is None:
                merged[candidate.segment_type] = SegmentCandidate(
                    segment_type=candidate.segment_type,
                    score=candidate.score,
                    reasons=list(candidate.reasons),
                )
                continue

            existing.score = max(existing.score, candidate.score)
            existing.reasons.extend(candidate.reasons)

        result = list(merged.values())
        for item in result:
            item.reasons = self._dedupe_keep_order(item.reasons)

        result.sort(key=lambda x: x.score, reverse=True)
        return result

    def _select_primary_segment(
        self,
        candidates: list[SegmentCandidate],
    ) -> SegmentType | None:
        if not candidates:
            return None

        best = candidates[0]
        if best.score < self._primary_threshold:
            return None

        return best.segment_type

    def _build_segment_buckets(
        self,
        *,
        allowed_segments: tuple[SegmentType, ...],
        block_recommendations: list[BlockSegmentRecommendation],
    ) -> list[SegmentBucket]:
        buckets: dict[SegmentType, SegmentBucket] = {
            segment: SegmentBucket(segment_type=segment)
            for segment in allowed_segments
        }

        for block_rec in block_recommendations:
            if block_rec.primary_segment is None:
                continue

            bucket = buckets[block_rec.primary_segment]
            bucket.block_refs.append(block_rec.block_ref)

            if block_rec.candidates:
                bucket.reasons.extend(block_rec.candidates[0].reasons)

        result: list[SegmentBucket] = []
        for segment in allowed_segments:
            bucket = buckets[segment]
            if not bucket.block_refs:
                continue
            bucket.reasons = self._dedupe_keep_order(bucket.reasons)
            result.append(bucket)

        return result

    def _collect_unmapped_block_refs(
        self,
        block_recommendations: list[BlockSegmentRecommendation],
    ) -> list[str]:
        return [
            block_rec.block_ref
            for block_rec in block_recommendations
            if block_rec.primary_segment is None
        ]

    def _build_warnings(
        self,
        *,
        doc_type: str,
        block_recommendations: list[BlockSegmentRecommendation],
        unmapped_block_refs: list[str],
    ) -> list[str]:
        warnings: list[str] = []

        found_segments = {
            block_rec.primary_segment
            for block_rec in block_recommendations
            if block_rec.primary_segment is not None
        }

        if SegmentType.ITEMS not in found_segments:
            warnings.append("items segment not found")

        if SegmentType.QUOTATION_META not in found_segments:
            warnings.append("quotation_meta segment not found")

        if unmapped_block_refs:
            warnings.append(f"{len(unmapped_block_refs)} block(s) are unmapped")

        return warnings

    @staticmethod
    def _dedupe_keep_order(values: list[str]) -> list[str]:
        seen: set[str] = set()
        result: list[str] = []
        for value in values:
            if value in seen:
                continue
            seen.add(value)
            result.append(value)
        return result
```
